### Задание №1.
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). Найдите их сумму. (на листочке)   
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

__Решение:__   
Суммой двух векторов является тоже вектор с координатами, равными суммам соответствующих координат суммируемых векторов:

$\normalsize (10,10,10) + (0,0,-10) = (10+0;10+0;10-10) = (10;10;0)$

```python
def vector_length(vector: list) -> float:
    """ Вычисление длины вектора
    """
    sq_sum = 0
    for i in vector:
        sq_sum += i ** 2
    
    return sq_sum ** 0.5
```

```python
a = [10, 10, 10]
b = [0, 0, -10]
c = [10, 10, 0]

vector_length(a), vector_length(b), vector_length(c)
```

### Задание №3.
Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
%config InlineBackend.figure_formats = ['svg']
```

```python
def build_circle(ax, r, x_0, y_0):
    """ Построение окружности с радиусом 'r' и центром в точке (x_0, y_0)
    """
    
    x = np.linspace(x_0 - r, x_0 + r, 201)
    y1 = y_0 + np.sqrt(r ** 2 - (x - x_0) ** 2)
    y2 = y_0 - np.sqrt(r ** 2 - (x - x_0) ** 2)
    
    ax.plot(x, y1)
    ax.plot(x, y2)
    ax.axes.set_aspect('equal')
    ax.set_title('circle')

    
def build_ellipse(ax, a, b, x_0, y_0):
    """ Построение эллипса с радиусами осей 'a' и 'b' и центром в точке (x_0, y_0)
    """
    
    x = np.linspace(x_0 - a, x_0 + a, 201)
    y1 = y_0 + np.sqrt(b ** 2 - b ** 2 * (x - x_0) ** 2 / a ** 2)
    y2 = y_0 - np.sqrt(b ** 2 - b ** 2 * (x - x_0) ** 2 / a ** 2)
    
    ax.plot(x, y1)
    ax.plot(x, y2)
    ax.axes.set_aspect('equal')
    ax.set_title('ellipse')


def build_hyperbola(ax, a, b, x_0, y_0):
    """ Построение гиперболы с радиусами действительной и мнимой осей 'a' и 'b' и центром в точке (x_0, y_0)
    """
    
    y = np.linspace(y_0 - 2 * b, y_0 + 2 * b, 201)
    x1 = x_0 + np.sqrt(a ** 2 * ( y- y_0) ** 2 / b ** 2 + a ** 2)
    x2 = x_0 - np.sqrt(a ** 2 * (y - y_0) ** 2 / b ** 2 + a ** 2)
    
    ax.plot(x1, y)
    ax.plot(x2, y)
    ax.set_title('hyperbola')
```

```python
fig, axes = plt.subplots(nrows=3)
fig.set_size_inches(5, 12)
axes = axes.flatten()

build_circle(axes[0], 2, 1, -1)
build_ellipse(axes[1], 3, 1, 1, -1)
build_hyperbola(axes[2], 3, 1, 0, 0)
```

### Задание №2.
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.

__Решение:__   
Пусть задано линейное преобразование из системы координат $\normalsize x0y$ в систему $\normalsize X0Y$:

$$\normalsize X = a_{11}x+a_{12}y+a_{13},$$
$$\normalsize Y = a_{21}x+a_{22}y+a_{23}.$$

Ортогональным будет преобразование при выполнении следующей системы уравнений:

$$\normalsize\begin{cases}a_{11}^2+a_{21}^2=1, \\ a_{12}^2+a_{22}^2=1, \\ a_{11}a_{12}+a_{21}a_{22}=0\end{cases}.$$

Расстояние между точками равно квадратному корню из суммы квадратов разностей соответствующих координат данных точек:

$$\normalsize r=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$
$$\normalsize R=\sqrt{(X_2-X_1)^2+(Y_2-Y_1)^2}$$

Докажем, что при ортогональном преобразовании $\normalsize r=R$, или $\normalsize r^2=R^2$, то есть $\normalsize (x_2-x_1)^2+(y_2-y_1)^2 = (X_2-X_1)^2+(Y_2-Y_1)^2$.

Для начала найдем разности координат:

$\normalsize X_2-X_1 = a_{11}x_2+a_{12}y_2+a_{13}-a_{11}x_1-a_{12}y_1-a_{13} = a_{11}(x_2-x_1)+a_{12}(y_2-y_1)$   
$\normalsize Y_2-Y_1 = a_{21}x_2+a_{22}y_2+a_{23}-a_{21}x_1-a_{22}y_1-a_{23} = a_{21}(x_2-x_1)+a_{22}(y_2-y_1)$

Тогда

$\normalsize R^2 = (X_2-X_1)^2+(Y_2-Y_1)^2 = \big(a_{11}(x_2-x_1)+a_{12}(y_2-y_1)\big)^2 + \big(a_{21}(x_2-x_1)+a_{22}(y_2-y_1)\big)^2 = a_{11}^2(x_2-x_1)^2 + a_{12}^2(y_2-y_1)^2 + 2a_{11}a_{12}(x_2-x_1)(y_2-y_1) + a_{21}^2(x_2-x_1)^2 + a_{22}^2(y_2-y_1)^2 + 2a_{21}a_{22}(x_2-x_1)(y_2-y_1) = (a_{11}^2+a_{21}^2)(x_2-x_1)^2 + (a_{12}^2+a_{22}^2)(y_2-y_1)^2 + 2(a_{11}a_{12}+a_{21}a_{22})(x_2-x_1)(y_2-y_1) = (*)$

Согласно соотношениям коэффициентов ортогонального преобразования получаем:

$\normalsize (*) = 1\cdot(x_2-x_1)^2 + 1\cdot(y_2-y_1)^2 + 2\cdot0\cdot(x_2-x_1)(y_2-y_1) = (x_2-x_1)^2 + (y_2-y_1)^2 = r^2$.

```python
def pol_to_dec(a, r):
    """ Перевод из полярных координат в декартовы
    """
    return r * np.cos(a), r * np.sin(a)


def dec_to_pol(x, y):
    """ Перевод из декартовых координат в полярные
    """
    r = np.sqrt(x ** 2 + y ** 2)
    a = 0.0
    if y > 0:
        a = np.arccos(x / r)
    elif y < 0:
        a = 2 * np.pi + np.arcsin(y / r) if x >= 0 else np.pi - np.arcsin(y / r)
    else:
        a = np.pi if x < 0 else a
    return a, r
```

1. Решите систему уравнений:

$\normalsize\begin{cases}y = x^2 – 1 \\ e^x + x(1 – y) = 1\end{cases}$

2. Решите систему уравнения и неравенства:

$\normalsize\begin{cases}y = x^2 – 1 \\ e^x + x(1 – y) - 1 > 0\end{cases}$

__(1) Решение:__   
Систему уравнений решим численным методом библиотеки `scipy` функцией `fsolve()`. Для этого сначала выразим переменную $y$ через $x$ в обоих уравнениях и построим графики получившихся функций на плоскости. Точки пересечений этих графиков и будут решениями нашей системы. Далее используем приближенные значения координат найденных точек в качестве начальных значений численного метода нахождения точных значений решения данной системы уравнений.

```python
x = np.linspace(-3, 5.5, 201)
f1 = x ** 2 - 1
f2 = (np.exp(x) - 1) / x + 1

plt.plot(x, f1)
plt.plot(x, f2)
```

Проанализировав график, определяем приблизительные значения координат точек пересечения: $\normalsize (-2, 1), (2, 5), (4, 15)$

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return y - x ** 2 + 1, x * (y - 1) - np.exp(x) + 1
```

```python
x1, y1 = fsolve(equations, (-2, 1))
x2, y2 = fsolve(equations, (2, 5))
x3, y3 = fsolve(equations, (4, 15))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```

__(1) Ответ:__   
Решения системы уравнений: $\normalsize (-1.58, 1.5), (2.62, 5.85), (4.2, 16.64)$

__(2) Решение:__   
Преобразуем данную систему уравнения и неравенства, выразив $y$ через $x$ в обоих выражениях и объединив в одно:

$\normalsize\displaystyle\begin{cases}y = x^2 – 1 \\ y < \displaystyle\frac{e^x-1}{x}+1\end{cases}\,\,\,\Rightarrow\,\,\,x^2 -1 < \frac{e^x-1}{x}+1\,\,\,\Rightarrow\,\,\,\frac{e^x-1}{x}-x^2+2>0$

Тогда решениями данного неравенства будут все значения $x$, при которых значения функции $\displaystyle f(x)=\frac{e^x-1}{x}-x^2+2$ будут больше 0.   
Построим график получившейся функции и по точкам пересечения с осью $0x$ найдем интервалы $x$, при которых функция положительна:

```python
f3 = (np.exp(x) - 1) / x + 2 - x ** 2

plt.plot(x, f3)
```
