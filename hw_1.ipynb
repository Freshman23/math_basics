### Задача №2.1.
Решите систему уравнений. Линейная или нелинейная это система? А каждое уравнение по отдельности?

$\normalsize\displaystyle \begin{cases} 3x–2y+5z=7 \\ 7x+4y–8z=3 \\ 5x–3y–4z=-12 \end{cases}$

__Решение:__   
Данная система уравнений является линейной, так как линейными являются каждое из уравнений в системе.   
Решим её методом Крамера:

$\normalsize\displaystyle \Delta = \begin{vmatrix} 3&–2&5 \\ 7&4&–8 \\ 5&–3&–4 \end{vmatrix} = -48-105+80-100-72-56 = -301$

$\normalsize\displaystyle \Delta_1 = \begin{vmatrix} 7&–2&5 \\ 3&4&–8 \\ 12&–3&–4 \end{vmatrix} = -112-45-192+240-168-24 = -301$

$\normalsize\displaystyle \Delta_2 = \begin{vmatrix} 3&7&5 \\ 7&3&–8 \\ 5&–12&–4 \end{vmatrix} = -36-420-280-75-288+196 = -903$

$\normalsize\displaystyle \Delta_3 = \begin{vmatrix} 3&–2&7 \\ 7&4&3 \\ 5&–3&–12 \end{vmatrix} = -144-147-30-140+27-168 = -602$

$\normalsize\displaystyle x = \frac{\Delta_1}{\Delta} = 1,\,\,\,\,\,y = \frac{\Delta_2}{\Delta} = 3,\,\,\,\,\,z = \frac{\Delta_3}{\Delta} = 2.$

__Ответ:__   
$\normalsize x = 1,\,\,\,y = 3,\,\,\,z = 2.$

### Задача №2.2.
Решите систему уравнений. Линейная или нелинейная это система? А каждое уравнение по отдельности?

$\normalsize\displaystyle \begin{cases} x^2+yx-9=0 \\ \displaystyle x-\frac{y}{5}=0 \end{cases}$

__Решение:__   
Данная система уравнений является нелинейной. В данной системе только второе уравнение является линейным, во втором - переменная x вступает во второй степени.   
Выразив одну переменную через другую во втором уравнении и подставив в первое, получим

$\normalsize\displaystyle \begin{cases} x^2+yx-9=0 \\ y=5x \end{cases}\,\,\,\Rightarrow\,\,\, x^2+5x\cdot x-9=0\,\,\,\Rightarrow\,\,\,6x^2=9\,\,\,\Rightarrow\,\,\,\begin{cases} x_{1,2}=\pm\displaystyle\frac{\sqrt{6}}{2} \\ y_{1,2}=\pm\displaystyle\frac{5\sqrt{6}}{2} \end{cases}$

__Ответ:__   
$\normalsize\displaystyle \left(\frac{\sqrt{6}}{2},\,\,\,\frac{5\sqrt{6}}{2}\right),\,\,\,\left(-\frac{\sqrt{6}}{2},\,\,\,-\frac{5\sqrt{6}}{2}\right)$
___

### Задача №3.
Площадь пола прямоугольной комнаты равна 48 м2,а его периметр равен 28 м. Найдите длину и ширину комнаты.

__Решение:__   
Формулы нахождения площади и периметра прямоугольника: $\normalsize S=xy,\,\,P=2(x+y)$.   
Зная их значения, составим систему уравнений и найдем стороны прямоугольника $x$ и $y$:

$\normalsize\displaystyle \begin{cases} xy=48 \\ 2(x+y)=28 \end{cases}\,\,\,\Rightarrow\,\,\,\begin{cases} xy=48 \\ y=14-x \end{cases}$

Подставив 2-ое уравнение в 1-ое, получим квадратное уравнение от одной переменной:

$\normalsize x(14-x)=48$   
$\normalsize x^2-14x+48=0$   
$\normalsize (x-7)^2-1=0$   
$\normalsize (x-7)^2=1$   
$\normalsize x-7=\pm1$   
$\normalsize x_1=8\,\,\,x_2=6$

Найдя значения $y=14-x$, получим две точки: $\normalsize (8,6), (6, 8)$. Что в данной задаче одно и то же решение.

__Ответ:__
длина - 8 м, ширина - 6 м.

### Задача №4.
Постройте на одном графике две кривые y(x) для функции двух переменных y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
%config InlineBackend.figure_formats = ['svg']
```

```python
y = lambda k, x: np.cos(k * x)
```

```python
x = np.linspace(-3*np.pi, 3*np.pi, 1000)
y1 = y(1, x)
y2 = y(2, x)
```

```python
plt.figure(figsize=(10, 4))

plt.plot(x, y1, label='cos(x)')
plt.plot(x, y2, label='cos(2x)')

plt.ylim(-1.1, 1.5)
plt.legend()
```
