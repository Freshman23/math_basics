### Задание №1.
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

__Решение:__   
Вероятность отдельного элементарного события (выпадения определенного стороны игральной кости): $p=1/6$.   
Так как события несовместны, по теореме сложения вероятностей вероятность события $A$ (выпадения 2 или 5 очков): $P(A)=p+p=1/3$

__Ответ:__ 1/3
___

### Задание №2.
Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

__Решение:__   
Вероятность отдельного элементарного события (выпадения определенного стороны игральной кости): $p=1/6$.   
Так как события несовместны и независимы, по теореме умножения вероятностей вероятность события $A$ (выпадения 2, а затем 5 очков при двух подряд подрасываниях): $P(A)=p\cdot p=1/36$

__Ответ:__ 1/36
___

### Задание №3.
Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости. Обратите внимание на порядок выпадения костей!

__Решение:__   
Событие $A$ выпадения 2 и 5 очков при двух подбрасываниях заключается в выпадении сначала 2, затем 5 очков, либо в выпадении 5, затем 2 очков. Обьединив теоремы сложения и умножения вероятностей, получим: $P(A)=1/6*1/6+1/6*1/6=1/18$

__Ответ:__ 1/18
___

### Задание №4.
На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

__Решение:__   
На отрезке AB отложим: точку D правее точки А на 9 см; точку E левее точки B на 15 см. Получим отрезок DE длиной: AD + BE - AB = 4 см. Тогда искомая вероятность равна: DE/AB=4/20=0.2

__Ответ:__ 0.2
___

### Задание №5.
Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882227?

__Решение:__    
Без всяких условностей вероятность конкретной последовательности из 10 цифр по 7 равна: $(1/10)^7 = 10^{-7}$.

Если же условится, что номер телефона не начинается с 0, то искомая вероятность равна: $(1/9)\cdot(1/10)^6= 1/9\cdot 10^{-6}$

__Ответ:__ $10^{-7}\,\,\,(1/9\cdot10^{-6})$
___

### Задание №6.
Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

__Решение:__   
Искомое количество вариантов $n$ - это количество перестановок из 9 по 2: $\displaystyle n=A_9^2=\frac{9!}{7!}=9\cdot8=72$.   
Так как правильный вариант только один $m=1$, то вероятность угадывания с первого раза: $P=m/n=1/72$.

__Ответ:__ 1/72
___

### Задание №7.
Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью  все грани этого куба будут белыми?

__Решение:__   
Найдем искомую вероятность как отношение числа перестановок кубиков ($m$), при которых каждый кубик находится в определенном положении (белый цвет наружу), к общему числу перестановок кубиков с учетом их положения ($n$).

Общее число перестановок кубиков: $A_{27}^{27} = 27!$.   
Также имеет значение как располагается кубик на своем месте. На каждой стороне 4 положения, сторон 6, тогда всего - 24 на один кубик. Всего размещений положений каждого кубика на одну перестановку: $24^{27}$.   
Тогда общее число перестановок кубиков с учетом их положения: $n=27!\cdot24^{27}$.

После покраски куба и разделения его на 27 кубиков получаем: 8 кубиков с 3мя белыми сторонами, 12 кубиков с 2мя белыми сторонами, 6 кубиков - с 1ой белой стороной и 1 кубик полностью черный. И при подсчете числа перестановок $m$ эти кубики меняют друг друга только из своей группы. Значит число перестановок будет равнятся произведению факториалов количеств кубиков в каждой группе.   
Также каждый из них удовлятворяет условиям только в нескольких из 24 определенных положениях, а точнее: 3 положения для кубиков с 3мя белыми сторонами, 2 положения - с 2мя белыми, 4 положения - с 1ой и все 24 положения для 1ого черного. Получаем количество размещений этих положений на одну перестановку: $3^8\cdot2^{12}\cdot4^6\cdot24$.   
Тогда число перестановок кубиков, при которых каждый кубик находится в определенном положении (белый цвет наружу): $m=8!\cdot12!\cdot6!\cdot1!\cdot3^8\cdot2^{12}\cdot4^6\cdot24$.

Получаем
$\displaystyle P=\frac{m}{n}=\frac{8!\cdot12!\cdot6!\cdot1!\cdot3^8\cdot2^{12}\cdot4^6\cdot24}{27!\cdot24^{27}} = \frac{8!\cdot12!\cdot6!}{27!}\cdot\frac{2^{27}\cdot3^9}{2^{81}\cdot3^{27}} = \frac{8!\cdot12!\cdot6!}{27!}\cdot\frac{1}{2^{54}\cdot3^{18}}\,\,\,\Rightarrow$

Выполнив некоторые сокращения и разложения на степени простыз чисел первой дроби, получим

$\displaystyle \Rightarrow\,\,\, \frac{1}{2^{56}\cdot3^{22}\cdot5^2\cdot7\cdot11\cdot13^2\cdot17\cdot19\cdot23} \approx 1,83\cdot10^{-37}$

__Ответ:__ $1,83\cdot10{-37}$
