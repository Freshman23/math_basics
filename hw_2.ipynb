### Задание №1.
Напишите уравнение параболы, проходящей через три точки (x,y):
(1,2), (3,10), (5,1)

__Решение:__   
Общее уравнение параболы имеет вид:
$$\normalsize y=ax^2+bx+c.$$

Подставив 3 имеющиеся точки, составим систему уравнений, состоящую из 3-х линейных уравнений, для нахождения 3-х неизвестных коэффициентов $a,b,c$.

$\normalsize\displaystyle \begin{cases}a+b+c=2 \\ 9a+3b+c=10 \\ 25a+5b+c=1\end{cases}$

Решим получившуюся систему методом Гаусса:   
- вычтем первое уравнение из второго и третьего:

$\normalsize\displaystyle \begin{cases}a+b+c=2 \\ 9a+3b+c=10 \\ 25a+5b+c=1\end{cases}\,\,\,\Rightarrow\,\,\, \begin{cases}a+b+c=2 \\ 8a+2b=8 \\ 24a+4b=-1\end{cases}$

- далее в получившейся системе домножим второе уравнение на 2 и вычтем его из третьего:

$\normalsize\displaystyle \begin{cases}a+b+c=2 \\ 8a+2b=8 \\ 24a+4b=-1\end{cases}\,\,\,\Rightarrow\,\,\, \begin{cases}a+b+c=2 \\ 8a+2b=8 \\ 8a=-17\end{cases}$

- посчитаем коэффициенты:

$\normalsize\displaystyle \begin{cases}a+b+c=2 \\ 8a+2b=8 \\ 8a=-17\end{cases}\,\,\,\Rightarrow\,\,\, \begin{cases}a=-2.125 \\ b=12.5 \\ c=-8.375\end{cases}$

__Ответ:__   
Искомая парабола имеет вид: $y=-2.125x^2+12.5-8.375$

### Задание №2.
Известно, что свежий огурец на 99% состоит из воды. Месяц назад взвесили мешок со свежими огурцами. Получилось, что мешок весил ровно 100 кг. Мешок убрали, а через месяц снова взвесили. Огурцы за это время усохли, и теперь вода составляет уже только 98% их веса. Сколько теперь (в кг) весят огурцы?

__Решение:__   
Составим уравнение где приравняем полностью обезвоженную массу огурцов до и после усушки:

$\normalsize (1-0.99)\cdot100=(1-0.98)\cdot x$,   

где $x$ - масса огурцов после усушки. Тогда

$\normalsize\displaystyle x=\frac{0.01\cdot100}{0.02}=50$

__Ответ:__   
50 кг

### Задание №3.
Определение логарифма. Решить уравнения:

1. $\normalsize 2^x=256$   

   $\normalsize x=\log_2{256}$   
   
   $\normalsize x=8$
___
2. $\normalsize 2^x=300$

   $\normalsize x=\log_2{300}$   
___
3. $\normalsize\displaystyle \log_8{2^{8x-4}}=4$   

   $\normalsize\displaystyle \log_{2^3}{2^{8x-4}}=4$   
   
   $\normalsize\displaystyle \frac{8x-4}{3}\cdot\log_2{2}=4$   
   
   $\normalsize\displaystyle \frac{8x-4}{3}=4$   
   
   $\normalsize\displaystyle x=2$
___
4. $\normalsize\displaystyle 3^{\displaystyle\log_9(5x-5)}=4$

   __ОДЗ:__ $5x-5>0\,\,\,\Rightarrow\,\,\,x>1$   
   
   $\normalsize\displaystyle 3^{\displaystyle\log_{3^3}(5x-5)}=4$   
   
   $\normalsize\displaystyle 3^{\displaystyle\log_3(5x-5)^\frac{1}{3}}=4$   
   
   $\normalsize\displaystyle (5x-5)^\frac{1}{3}=4$
   
   $\normalsize\displaystyle 5x-5=64$
   
   $\normalsize\displaystyle x=13.8$
___
5. $\normalsize\displaystyle x^{\displaystyle\log_3x+1}=9$
   
   __ОДЗ:__ $x>0$
   
   Прологарифмируем по основанию обе части уравнения:
   
   $\normalsize\displaystyle \log_3{x^{\displaystyle\log_3x+1}}=\log_3{9}$
   
   $\normalsize\displaystyle (\log_3{x}+1)\cdot\log_3{x}=2$
   
   Выполним замену переменной: $\normalsize t=\log_3{x}\,\,\,\Rightarrow\,\,\,x=3^t$
   
   $\normalsize (t+1)t=2$
   
   $\normalsize t^2+t-2=0$
   
   $\normalsize (t+0.5)^2-2.25=0$
   
   $\normalsize (t+0.5)^2=2.25$
   
   $\normalsize t+0.5=\pm1.5$
   
   $\normalsize t_1=1\,\,\,t_2=-2$
   
   Тогда   
   $\normalsize x_1=3\,\,\,x_2=\displaystyle\frac{1}{9}$
___
