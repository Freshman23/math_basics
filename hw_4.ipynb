### Задание №1.
Решите уравнение:   
$\displaystyle\normalsize \frac{\sin(x)}{x}=0$

__Решение:__   
Определяем область допустимых значений: $\normalsize x\neq0$.   
Тогда задача сводится к решению уравнения: $\normalsize \sin(x)=0$ с учетом ОДЗ.

$\normalsize x=\arcsin(0)$   
$\normalsize x=\pi n, n\in\mathbb Z$   

C учетом ОДЗ получаем   
$\normalsize x=\pm\pi n, n\in\mathbb N$

__Ответ:__   
$\normalsize x=\pm\pi n, n\in\mathbb N,\,\,\big(\mathbb N=\{1,2,3,4,...\}\big)$

### Задание №2.
Даны три прямые $y=k_1x+b_1$, $y=k_2x+b_2$, $y=k_3x+b_3$. Как узнать, пересекаются они в одной точке или нет?

__Решение:__   
Пересечение трёх прямых в одной точке означает равенство $x$ и $y$, поэтому для начала составим систему уравнений, попарно приравняв выражениея прямых:

$\normalsize\displaystyle \begin{cases}k_1x+b_1 = k_2x+b_2 \\ k_2x+b_2 = k_3x+b_3 \\ k_1x+b_1 = k_3x+b_3\end{cases}$

При чём достаточно выполения двух равенств из трёх.   
Теперь выразим $-x$ через коэффициенты и приравняем их:

$\normalsize\displaystyle \begin{cases}-x = \displaystyle\frac{b_1-b_2}{k_1-k_2} \\ -x = \displaystyle\frac{b_2-b_3}{k_2-k_3} \\ -x = \displaystyle\frac{b_1-b_3}{k_1-k_3}\end{cases}\,\,\,\Rightarrow\,\,\,\frac{b_1-b_2}{k_1-k_2} = \frac{b_2-b_3}{k_2-k_3} = \frac{b_1-b_3}{k_1-k_3}$

Для определения факта пересечения трёх прямых в одной точке достаточно выполнения одного из равенств.

__Ответ:__   
$\normalsize\displaystyle \frac{b_1-b_2}{k_1-k_2} = \frac{b_2-b_3}{k_2-k_3} = \frac{b_1-b_3}{k_1-k_3}$

### Задание №3.

### Задание №4.
Решите аналитически и потом численно уравнение, зависящее от параметра а: $\sin(аx)=0$ при условии: $0.01<a<0.02, 100<х<500$.

__Решение:__   

$\normalsize\sin(аx)=0$   
$\normalsize аx=arcsin(0)$   
$\normalsize аx=\pi n, n\in\mathbb Z$   
$\normalsize\displaystyle x=\frac{\pi n}{a}, n\in\mathbb Z$

Рассмотрим попадание $x$ в заданный диапазон $(100, 500)$ при разных значениях $n$ и заданном диапазоне $a\in(0.01,0.02)$:   
- при $n\le0$:   
  Получаемое решение $x(a)\le0$ при изменении $a$ от $0.01$ до $0.02$ и не входит в заданный диапазон.


- при $n=1$: $\displaystyle x(a)=\frac{\pi}{a}$   
  И при изменении $a$ от $0.01$ до $0.02$ - $x$ изменяется от $314$ до $157$ (от $100\pi$ до $50\pi$). Входит в заданный диапазон.   
  Тогда одним из решений уравнения будет:   
  $x(a)=\pi/a$, при $a\in(0.01; 0.02)$.


- при $n=2$: $\displaystyle x(a)=\frac{2\pi}{a}$   
  И при изменении $a$ от $0.01$ до $0.02$ - $x$ изменяется от $628$ до $314$ (от $200\pi$ до $100\pi$). Получаем, что только часть значений входит в заданный диапазон.   
  Найдем левую границу диапазона аргумента $a$: $500=2\pi/a_l\,\,\,\Rightarrow\,\,\,a_l=2\pi/500\approx0.013$.   
  Ещё одним решением уравнения будет:   
  $x(a)=2\pi/a$, при $a\in(0.013; 0.02)$.


- при $n=3$: $\displaystyle x(a)=\frac{3\pi}{a}$   
  И при изменении $a$ от $0.01$ до $0.02$ - $x$ изменяется от $942$ до $471$ (от $300\pi$ до $150\pi$). Также получаем, что только часть значений входит в заданный диапазон.   
  Найдем левую границу диапазона аргумента $a$: $500=3\pi/a_l\,\,\,\Rightarrow\,\,\,a_l=3\pi/500\approx0.019$.   
  Ещё одним решением уравнения будет:   
  $x(a)=3\pi/a$, при $a\in(0.019; 0.02)$.


- при $n\ge4$:   
  Получаемое решение $x(a)\ge628$ при изменении $a$ от $0.01$ до $0.02$ и не входит в заданный диапазон.


Изобразим решение на графике:

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
%config InlineBackend.figure_formats = ['svg']
```

```python
x_func = lambda a, n: np.pi * n / a
```

```python
a_1 = np.array([0.01, 0.02])
x_1 = x_func(a_1, 1)
a_2 = np.array([2 * np.pi / 500, 0.02])
x_2 = x_func(a_2, 2)
a_3 = np.array([3 * np.pi / 500, 0.02])
x_3 = x_func(a_3, 3)

plt.figure(figsize=[11, 8])

plt.title('$x(a)=\pi n/a$', size='large')
plt.xlabel('a', size='large')
plt.ylabel('x(a)', labelpad=15, rotation='horizontal', size='large')
plt.xlim([0.008, 0.022])
plt.ylim([0, 600])
plt.xticks(list(plt.xticks()[0]) + [round(x, 5) for x in [0.004 * np.pi, 0.006 * np.pi]], rotation='vertical')
plt.yticks(list(plt.yticks()[0]) + [round(y, 1) for y in [50 * np.pi, 100 * np.pi, 150 * np.pi]])

plt.plot(a_1, x_1, 'o-r', lw=3, label='n = 1')
plt.plot(a_2, x_2, 'o-r', lw=3, label='n = 2')
plt.plot(a_3, x_3, 'o-r', lw=3, label='n = 3')
plt.text(0.0192, 450, '$n=3$')
plt.text(0.0162, 360, '$n=2$')
plt.text(0.0142, 210, '$n=1$')
plt.plot([0.009, 0.021], [100, 100], '--', c='grey')
plt.plot([0.009, 0.021], [500, 500], '--', c='grey')
plt.plot([0.01, 0.01], [50, 550], '--', c='grey')
plt.plot([0.02, 0.02], [50, 550], '--', c='grey')
plt.show()
```
